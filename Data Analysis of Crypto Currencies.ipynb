The objective of this project is the analysis of multiple cryptocurrencies to get an overview of the cryptocurrency market. Then the statistical analysis of bitcoin historical prices carried out with the help of Seaborn and Matplotlib library of Python

```python
# Importing relevant libraries
import pandas as pd
import numpy as np
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
# Fetching data from an excel file
data = pd.read_excel('1000_crypto_DA.xlsx')
```

```python
# Showing the name of columns, their data types, and range
print(data.info(),data.columns)
```

```python
# Reindexing to columns
data.columns=['Coin_name', 'Price($)', 'Change_1h(%)', 'Change_24h(%)', 'Change_7d(%)', 'Volume($)', 'Market cap($)']
print(data.info(),data.columns)
```

```python
# Rows with null values
print(data[data.isnull().T.any()])
```

```python
# Dropping all the rows with null values
data=data.dropna()
#Printing the new data
data
```

```python
# Getting the information of Columns
print(data.info(),data.columns)
```

```python
# Finding the highest and lowest from all the CryptoCurrencies according to their Market Capitalization,
# 1 hour change, 24 hour change, 7 day change, and Volume

maxMarketCap=data.loc[data['Market cap($)']==data['Market cap($)'].max(),['Coin_name','Market cap($)']]
minMarketCap=data.loc[data['Market cap($)']==data['Market cap($)'].min(),['Coin_name','Market cap($)']]
highest_gainer_1hour = data.loc[data['Change_1h(%)']==data['Change_1h(%)'].max(),['Coin_name', 'Change_1h(%)']]
biggest_loser_1hour = data.loc[data['Change_1h(%)']==data['Change_1h(%)'].min(),['Coin_name', 'Change_1h(%)']]
highest_gainer_24hour = data.loc[data['Change_24h(%)']==data['Change_24h(%)'].max(),['Coin_name', 'Change_24h(%)']]
biggest_loser_24hour = data.loc[data['Change_24h(%)']==data['Change_24h(%)'].min(),['Coin_name', 'Change_24h(%)']]
highest_gainer_7d = data.loc[data['Change_7d(%)']==data['Change_7d(%)'].max(),['Coin_name', 'Change_7d(%)']]
biggest_loser_7d = data.loc[data['Change_7d(%)']==data['Change_7d(%)'].min(),['Coin_name', 'Change_7d(%)']]
crypto_largest_volume = data.loc[data['Volume($)']==data['Volume($)'].max(),['Coin_name', 'Volume($)']]
crypto_smallest_volume = data.loc[data['Volume($)']==data['Volume($)'].min(),['Coin_name', 'Volume($)']]
print('Currency with highest Market Cap', maxMarketCap, 'Currency with Lowest Market Cap', minMarketCap,
      'Highest gainer of 1 hour', highest_gainer_1hour, 'Biggest loser of 1 hour', biggest_loser_1hour, 
      'Highest gainer of 24 hour', highest_gainer_24hour, 'Biggest loser of 24 hour', biggest_loser_24hour,
      'Highest gainer of 7 days', highest_gainer_7d, 'Biggest loser of 24 hour', biggest_loser_7d,
      'Crypto with highest volume(24 hours)', crypto_largest_volume, 'Crypto with lowest volume(24 hours)', crypto_smallest_volume,
      sep='\n')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
x=pd.read_csv('Bitcoin USD (Historical Data)HistoricalData12.16-0..43.csv')
print(x.info(),x.columns)
x=x[['Date', 'Market Cap', 'Volume', 'Open', 'Close']]
print(x[x.isnull().T.any()]) #test whether has N/A
print(x)
x=x.dropna()  #delete the row having N/A (the first row)
print(x.head())
```

```python
print("This document contains Bitcoin's daily price data from {} to {},totalling {} numbers of data".format(x.iloc[-1,0],x.iloc[0,0],len(x['Open'])))
```

```python
print(x.info(),x.columns)
```

```python
maxMarketCap=x.loc[x['Market Cap']==x['Market Cap'].max(),['Date','Market Cap']]
minMarketCap=x.loc[x['Market Cap']==x['Market Cap'].min(),['Date','Market Cap']]
print('maxMarketCap',maxMarketCap,'minMarketCap',minMarketCap,sep='\n')
```

```python
x['AvePrice']=(x['Open']+x['Close'])/2
maxAvePrice=x.loc[x['AvePrice']==x['AvePrice'].max(),['Date','AvePrice']]
minAvePrice=x.loc[x['AvePrice']==x['AvePrice'].min(),['Date','AvePrice']]
print('maxAvePrice',maxAvePrice,'minAvePrice',minAvePrice,sep='\n')
```

```python
frequencyData=pd.cut(x['AvePrice'],list(range(15500,21501,1000)),right=False)
AvePriceFrequency=frequencyData.value_counts(sort=False)
print('Frequency',AvePriceFrequency,'Frequency Distribution',frequencyData.value_counts(sort=False,normalize='True'),sep='\n')
#Average price interval distribution, frequency distribution and positive probability distribution
```

```python
%matplotlib inline
x['AvePrice'].plot.density() #Distribution density map
y=seaborn.distplot(x['AvePrice'])
plt.xlabel('price')
plt.title('Distribution Plot for Average Price')
y=seaborn.displot(x['AvePrice'])
plt.xlabel('price')
plt.title('Distribution Plot for Average Price')
x['AvePrice'].describe()
```

```python
import matplotlib.ticker as ticker
seaborn.catplot(x='Market Cap',y='AvePrice',data=x)
plt.xlabel('Market Cap')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(6)) #Display only part of horizontal ordinates. 6 is the Spacing distance
plt.xticks(rotation=-90)
plt.ylabel('Price')
#Scatter plot
```

```python
plt.plot(x['Date'],x['Market Cap'],label='Market Cap',color='b')
plt.xlabel('Date')
plt.title('Market Cap')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(6))
plt.xticks(rotation=-90)
plt.show()

plt.plot(x['Date'],x['AvePrice'],label='Price',color='green')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(6))
plt.xlabel('Date')
plt.title('Price')
plt.xticks(rotation=-90)
plt.show()
```

```python
print(x.corr(),'\n')
import statsmodels.api as sm #Study the linear relationship between market cap and price
result=sm.OLS(x['Market Cap'],x['AvePrice']).fit()
print(result.summary())
```

```python
Y=pd.DataFrame() #Autoregression of stock price data. today and yesterday
Y['todayPrice']=x['AvePrice'][:58].reset_index(drop=True)
Y['yesterdayPrice']=x['AvePrice'][1:].reset_index(drop=True)
print(Y.info())
result=sm.OLS(Y['todayPrice'],Y['yesterdayPrice']).fit()
print(result.summary())
```
