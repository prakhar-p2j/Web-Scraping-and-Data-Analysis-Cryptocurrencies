The objective of this project is to fetch the data of crypto-currencies from 'www.coingecko.com/' and arrange them in a dataframe using Pandas libary of Python. The first code below fetches data of 1000 Cryptocurrencies from 100 Pages of the website. And the Second code fetches the price history of Bitcoin.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Making Empty Lists

name = []
price = []
change_1h = []
change_24h = []
change_7D = []
volume = []
marketcap = []

# Using 'For' loop to scrap multiple pages from 'https://www.coingecko.com/'. 
# We are scraping 10 pages so keeping 'range' till 10.

for i in range (1,11):
    # website is a variable in which web page is stored
    website = 'https://www.coingecko.com/?page=' + str(i)
    
    # request to get website
    response = requests.get(website)
    
    # Checked if the website is properly fetched or not with this code - 'response.status_code'. 
    # And it gives the output - '200'. which means the web page properly fetched
    
    # creating soup object to parse HTML page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Securities is a variable in which whole web-scraped table is stored
    securities = soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
    
    # Creating a 'For' loop to get individual elements from 'securities' data
    for element in securities:
        # fetching names of the crypto-currencies in the list named 'name'
        try:
            name.append(element.find('span', {'class':'lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
        except:
            name.append('n/a')

        # fetching prices of the crypto-currencies in the list named 'price'
        try:
            price.append(float(element.find('td',{'class': 'td-price'}).get_text().strip()[1:].replace(',','')))
        except:
            price.append('n/a')

        # fetching one hour price change of the crypto-currencies in the list named 'change_1h'
        try:
            change_1h.append(float(element.find('td',{'class': 'td-change1h'}).get_text().strip().replace('%','')))
        except:
            change_1h.append('n/a')

        # fetching 24 hour price change of the crypto-currencies in the list named 'change_24h'
        try:
            change_24h.append(float(element.find('td',{'class': 'td-change24h'}).get_text().strip().replace('%','')))
        except:
            change_24h.append('n/a')

        # fetching 7 Days price change of the crypto-currencies in the list named 'change_7D'
        try:
            change_7D.append(float(element.find('td',{'class': 'td-change7d'}).get_text().strip().replace('%','')))
        except:
            change_7D.append('n/a')

        # fetching 24 hour volume of the crypto-currencies in the list named 'volume'
        try:
            volume.append(float(element.find('td',{'class': 'td-liquidity_score lit text-right col-market'}).get_text().strip()[1:].replace(',','')))
        except:
            volume.append('n/a')

        # fetching market capitalization of the crypto-currencies in the list named 'marketcap'
        try:
            marketcap.append(float(element.find('td',{'class': 'td-market_cap cap col-market cap-price text-right'}).get_text().strip()[1:].replace(',','')))
        except:
            marketcap.append('n/a')    
            
# Creating Data Frames to properly arrange the data with the help of 'Pandas' library

element_df = pd.DataFrame({'Coin_name' : name, 'Price($)': price, 'Change_1h(%)': change_1h, 'Change_24h($)': change_24h, 
                           'Change_7d(%)': change_7D, 'Volume($)': volume, 'Market cap($)': marketcap })

# Concerting the data into an Excel file
element_df.to_excel('1000_crypto_DA.xlsx', index = False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn

time=datetime.now()
currentDateAndTime='{}.{}-{}..{}'.format(time.month,time.day,time.hour,time.minute)
coin='https://www.coingecko.com/en/coins/bitcoin/historical_data?start_date=2022-03-01&end_date=2022-11-30#panel'
x=requests.get(coin)
soup=BeautifulSoup(x.text,'lxml')
head=soup.find('h1').get_text()
table=soup.find('table', {'class':"table table-striped text-sm text-lg-normal"})
# print(table) #return 'tag'
headTitle=table.find('thead').get_text()
HeadTitleList=headTitle.strip().split('\n')

headTitle=headTitle.strip().split('\n') #a list
#print(len(table.find('tbody'))) #63
DataList=[]
for i in table.find('tbody'):
    if i=='\n':
        continue
    DailyData=i.get_text().strip().split()
    DataList.append(DailyData)
file2=pd.DataFrame(DataList,columns=HeadTitleList)
file2['Market Cap']=file2['Market Cap'].str.replace(',','').str.replace('$','')
file2['Volume']=file2['Volume'].str.replace(',','').str.replace('$','')
file2['Open']=file2['Open'].str.replace(',','').str.replace('$','')
file2['Close']=file2['Close'].str.replace(',','').str.replace('$','').str.replace('N/A','')
file2['Market Cap']=file2['Market Cap'].astype(float)
file2['Volume']=file2['Volume'].astype(float)
file2['Open']=pd.to_numeric(file2['Open'])
file2['Close']=pd.to_numeric(file2['Close'])
file2.to_excel('{}HistoricalData{}.xlsx'.format(head.strip(),currentDateAndTime))
file2.to_csv('{}HistoricalData{}.csv'.format(head.strip(),currentDateAndTime))
```
